```python
import json

# Method 1: Read line by line and catch the error
with open("indeterminant/predictions-bengali-generated-translated.jsonl", 'r', encoding='utf-8') as f:
    for line_num, line in enumerate(f, start=1):
        try:
            json.loads(line)
        except json.JSONDecodeError as e:
            print(f"Error on line {line_num}: {e}")
            print(f"Content: {line[:200]}...")  # Print first 200 chars
            break
```
