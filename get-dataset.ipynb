```python
import json

filtered_df = df[df['ID'].isin(data)]

reformatted = {}

for _, row in filtered_df.iterrows():
    question_id = row['ID']

    answers = []
    for annotation in row['annotations']:
        # Use the original language answers
        answers.extend(annotation['answers'])

    reformatted[question_id] = {
        "english_question": row['question'],
        "english_answer": answers,
        "original_english_question": row['en_question'],
    }

# Save to JSON file
with open('annotated_cultural_data_assam.json', 'w', encoding='utf-8') as f:
    json.dump(reformatted, f, ensure_ascii=False, indent=2)
```
